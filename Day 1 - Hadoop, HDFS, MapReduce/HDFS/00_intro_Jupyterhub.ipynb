## Getting started with Jupyter notebooks

A Jupyter notebook is an interactive Web application that allows editing and running Jupyter notebooks (in `ipynb` format) documents. The kernel is a runtime environment that provides programming language support for the Jupyter Notebook application. 



### What is a kernel? 

![Select kernel](kernel.png)

A kernel is essentially a _computation engine_ that is able to interpret your commands. A Python 3 kernel recognizes Python 3 commands. 
Kernels for many programming languages are available and can have some extra configuration and features, but for now you shouldn't be concerned with that for now unless you want to develop your own kernel.

You can execute Python code in any code cell.

```python
a = [4, 3, 5, 2, 1]
sorted(a)
```

**Python example 3: plot something**

Note: to view the plot in the notebook you need to enter the directive `%matplotlib inline`.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 4*np.pi, 0.1)   
y = np.cos(x)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel='x', ylabel='cos(x)', title='cosine')
ax.grid()

plt.show()
```

## Running bash commands

Underlying the Python interpreter there's a `bash` shell that allows you to run commands on the host machine. To run shell commands you need to _escape_ the Python interpreter by:

 - using an exclamation mark (`!`)  at the beginning of the command if the command fits in a single line
 
 - using the line `%%bash` at the beginning of a cell (whole first line in the cell) if you want to execute a multi-line bash script
 
**Bash example 1: list files**

For example, you can run the `ls` command to list the files in your current directory (the directory where the notebook resides):

```python
!ls 
```

**Bash example 2: show environment variable**

```python
!echo "My username is: "$USER
```

**Bash example 3: multi-line script**

List all files named `*ipynb`

```python
%%bash
cd ~/hadoop_training/HDFS
PATTERN="*.ipynb"
for i in $PATTERN; do
    echo $i
done
```

**Bash example 4: environment variables**

You have some predefined environment variables (given by the bash + notebook envirmonment). Any newly defined environment variable lives only within the cell where it's defined.

```python
%%bash
# this is a bash comment
export PATH=/my/newpath
echo $PATH
```

But in another cell your change in the bash environment variable is lost.

```python
!echo $PATH
```
